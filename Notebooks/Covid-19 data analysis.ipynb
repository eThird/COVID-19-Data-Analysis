```python
#step 3 now drop the unnessaryd ata from the sheet or  our columns from the first sheet 
covid_df.drop(['Sno','Time','ConfirmedIndianNational','ConfirmedForeignNational'], inplace= True, axis=1)
covid_df.head(10)
```

```python
#maing the format of date same at all
covid_df['Date']= pd.to_datetime(covid_df['Date'], format='%Y-%m-%d')
covid_df.head(5)
```

```python
#find total no of active cases - active cases = total numbr of confirmend cases -sum of(cured cases +deaths reported)
covid_df['Active_Cases']= covid_df['Confirmed']-( covid_df['Cured'] + covid_df['Deaths'])
covid_df.head()
```

```python
#now summing the  deacths and confired caes for each state and union teritories 
statewise=pd.pivot_table(covid_df,values=['Confirmed','Cured', 'Deaths'], index='State/UnionTerritory',aggfunc= max)
statewise.head()
```

```python
#findin teh recovery rate which is totalnumber of cures cases/total number of connfermed cases *100
statewise['Recovery Rate']= (statewise['Cured']*100)/ statewise['Confirmed']
```

```python
#mortality rate 
statewise['Mortality Rate']= (statewise['Deaths']*100)/ statewise['Confirmed']
```

```python
#sort the values accouding to confirmed cases 
statewise= statewise.sort_values(by ='Confirmed', ascending = False)
```

```python
# getting the output in better way
statewise.style.background_gradient(cmap="cubehelix")
```

```python
# TOP 10  active cases by state 
top_10_active_cases= covid_df.groupby(by='State/UnionTerritory').max()[['Active_Cases','Date']].sort_values(by=['Active_Cases'], ascending =False).reset_index()
```

```python
# TOP 10  active cases by state 
top_10_active_cases= covid_df.groupby(by='State/UnionTerritory').max()[['Active_Cases','Date']].sort_values(by=['Active_Cases'], ascending =False).reset_index()
#now plot those data we will be using matplot lib  for that
fig=plt.figure(figsize=(16,9))
plt.title('Top 10 states wit most active cases')
ax= sns.barplot(
    data= top_10_active_cases.iloc[:10], 
    y='Active_Cases', x='State/UnionTerritory', 
    linewidth=1 ,edgecolor='red',
    palette="viridis"  # Try "coolwarm", "plasma", "magma", etc.
               
)


plt.xlabel('States')
plt.ylabel('Total Active Cases')
plt.show()

```

```python
#top states with highest deaths 
top_10_deaths = covid_df.groupby(by='State/UnionTerritory').max()[['Deaths','Date']].sort_values(by= ['Deaths'],ascending= False).reset_index()

#now plot graphs
fig=plt.figure(figsize=(18,10))
plt.title('top 10 states with highest number of deaths')
#now giving the values of graph in side
ax=sns.barplot(
    data= top_10_deaths.iloc[:10],
    x='State/UnionTerritory', y='Deaths',
    linewidth=2 , edgecolor='black',
    #for giving colourfuk outlook
    hue='State/UnionTerritory',  # Assigning `hue`
  
    palette='coolwarm',
    
)
```
